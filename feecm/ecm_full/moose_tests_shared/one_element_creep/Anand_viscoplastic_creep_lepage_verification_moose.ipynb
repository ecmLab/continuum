```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
homedir = '/srv/repo1/srinath/Projects/hyperelastic_viscoplastic_tests/one_element_creep'
temp = [198,298,398]
str_rate = [4e-5, 3e-4, 3e-3, 2e-2]
s_rate = ['4e-5', '3e-4', '3e-3', '2e-2']
base_file = 'Li_one_elem_t_'
fnames = [base_file + x for x in s_rate]
```

```python
os.chdir(homedir)
ax1 = []
fig, ax1 = plt.subplots(1, len(temp), figsize=(18,16))
for i,s in enumerate(temp):
    ax1[i].set_prop_cycle(None)
    color = next(ax1[i]._get_lines.prop_cycler)['color']
    os.chdir(str(s))
    for f in fnames:
        f1 = f + '_out.csv'
        df = pd.read_csv(f1)
        ax1[i].plot(df['strain_zz'], df['stress_zz']/1e6)
    os.chdir('../')
```
