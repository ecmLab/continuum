```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('tension_out.csv')
```

```python
df['engineering_strain']  = df['u_y']/15.5e-3
df['engineering_stress'] = df['stress_yy']/(1.0 + df['engineering_strain'])
df['engineering_stress2'] = df['von_mises']/(1.0 + df['engineering_strain'])
```

```python
plt.plot(df['engineering_strain'], df['engineering_stress2']/1e6)
plt.xlim([0,0.5])
```

```python
plt.plot(df['engineering_strain'], df['engineering_stress']/1e6)
plt.xlim([0,0.01])
plt.ylim([0,1.3])
```
